```python
import requests
from bs4 import BeautifulSoup
site=requests.get("https://www.boraautocare.co.ke/")

print(site)
print(site.text)

# Parsing the HTML
soup = BeautifulSoup(site.text, 'html.parser')
print(soup.prettify())
```

```python
s = soup.find('div', class_='container')
content = s.find_all('a')

print(content)
```
